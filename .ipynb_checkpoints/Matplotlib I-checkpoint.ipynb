## Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4]);
```

```python
#Alternative plt.plot() plt.show()
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y);
```

```python
fig_var = plt.figure()
axes = fig_var.add_subplot()
plt.show()
```

## 2nd Method


```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

## 3rd Method Recommended 

```python
fig,ax = plt.subplots()
ax.plot(x,[10,15,25,40]);
type(fig),type(ax)
```

```python
dep_var = [20,40,60,80,100]
indep_var = [2,4,6,8,10]
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(indep_var,dep_var);
ax.set(title="Simple Graph", xlabel="Age/years", ylabel="Money/£");
```

## Making Figures out of NumPy Arrays

```python
x_arr = np.linspace(0,10,100)
```

```python
figur, axi = plt.subplots(figsize=(5,5))
axi.scatter(x_arr,np.exp(x_arr))
```

```python
nut_butter_prices = {"Almond Butter":10,
                     "Cashew Butter":14,
                     "Peanut Butter":7}
figr,axe = plt.subplots()
axe.bar(nut_butter_prices.keys(),nut_butter_prices.values())
axe.set(title="Youssef's Nut Butter Prices",xlabel="Nut Butter Type",ylabel="Price/£");
```

## Horizontal Bar Chart

```python
figrh,axeh = plt.subplots()
axeh.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
axe.set(title="Youssef's Nut Butter Prices",xlabel="Nut Butter Type",ylabel="Price/£");
```
