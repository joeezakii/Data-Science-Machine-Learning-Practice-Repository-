```python
car_sales['Price'] = car_sales['Price'].str.replace('$', '', regex=False)
car_sales['Price'] = car_sales['Price'].str.replace(',', '', regex=False)
car_sales['Price'] = car_sales['Price'].astype(float).astype(int)
```

```python
car_sales.plot(x='Make', y='Price')
```
