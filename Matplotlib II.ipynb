```python
heart_table = pd.read_csv("heart.csv")
heart_table.head(23)
```

```python
heart_table["Age"].plot(kind="hist", color="yellow");
```

## Which Method should I use? Matplotlib OO method or Pyplot?
. When plotting something quickly, use the pyplot method.

. When plotting something more advanced, use tbe OO method.

```python
over55 = heart_table[heart_table["Age"] > 55]
over55
cholesterol_0 = heart_table[heart_table["Cholesterol"] == 0]
cholesterol_0
```

```python
over55.plot(kind="scatter", x="Age", y="Cholesterol", c="HeartDisease");
```

## OO Method mixed with pyplot

```python
figure,axis = plt.subplots(figsize=(10,6))
over55.plot(kind="scatter", x="Age", y="Cholesterol", 
            c="HeartDisease",ax=axis
           );
axis.set_xlim([55,80]);
```

## OO Method from scratch

```python
fig,axi = plt.subplots(figsize=(7,5))
scatter_grph = axi.scatter(x=over55["Age"], y= over55["Cholesterol"],c= over55["HeartDisease"]);
axi.set(title="Heart Disease Scatter Graph for People Over the Age of 55", xlabel="Age/Years", ylabel="Cholesterol/(mg/dL)");
scatter_grph;
axi.set_xlim([55,80])
axi.legend(*scatter_grph.legend_elements(),title="Heart Disease");
axi.axhline(over55["Cholesterol"].mean(), linestyle="--", color="black", alpha=0.6);
scatter_grph;
```

## 2 Subplots between Age, Cholesterol & RestingBP (Resting Blood Pressure)

```python
figure, (axis0,axis1) = plt.subplots(nrows=2, ncols=1, figsize=(9,11), sharex=True)
scatter1 = axis0.scatter(x=over55["Age"],y=over55["Cholesterol"],c=over55["HeartDisease"],)
axis0.set(title="Heart Disease Scatter Graph for People Over 55 Years of Age (Cholesterol Readings)",xlabel="Age/Years",ylabel="Cholesterol/(mg/dL)")
scatter2 = axis1.scatter(x=over55["Age"], y=over55["RestingBP"],c=over55["HeartDisease"])
axis1.set(title="Heart Disease Scatter Graph for People Over 55 Years of Age (Resting Blood Pressure Readings) ",xlabel="Age/Years",ylabel="RestingBP/(mm/Hg)")
axis0.legend(*scatter1.legend_elements(),title="Heart Disease (Cholesterol Readings)");
axis0.axhline(over55["Cholesterol"].mean(), linestyle="--", color="black", alpha=0.6);
axis1.legend(*scatter2.legend_elements(),title="Heart Disease (Resting Blood Pressure)")
axis1.axhline(over55["RestingBP"].mean(), linestyle=":", color="green", alpha=0.9)
figure.suptitle("Heart Disease Readings for People Over The Age of 55", fontsize=20, fontweight="bold")
plt.show()
```
