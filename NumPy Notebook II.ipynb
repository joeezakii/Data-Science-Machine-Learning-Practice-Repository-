## Dot Product Example (Nut Butter Sales)

```python
np.random.seed(22)
sales_amounts = np.random.randint(12,size=(5,3))
sales_amounts
```

```python
df = pd.DataFrame(sales_amounts,index=["Monday","Tuesday","Wednesday","Thursday","Friday"])
```

```python
df
```

```python
df.columns = ["Peanut Butter", "Almond Butter", "Cashew Butter"]
```

```python
df
```

```python
price = np.random.randint(3,11,size=(1,3))
```

```python
price
```

```python
prices_df = pd.DataFrame(price, index=["Price"],columns=["Peanut Butter", "Almond Butter", "Cashew Butter"])
```

```python
prices_df
```

```python
total_sales = price.dot(sales_amounts.T)
```

```python
total_sales
```

```python
daily_sales = prices_df.dot(sales_amounts.T)
daily_sales.columns = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
daily_sales
```
